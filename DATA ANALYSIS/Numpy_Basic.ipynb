# 1.Create numpy array of size (2, 3) filled with zeros

```python
import numpy as np
np.zeros((2, 3))
```

# 2.Create numpy array of size(5, 1) filled with ones

```python
np.ones((5, 1), dtype=int)
```

# 3.Combine two numpy array (3, 2) + (2, 2) => (5, 2)

```python
a = np.array([[1, 2], [3, 4], [5, 6]])
b = np.array([[7, 8], [9, 10]])
np.vstack((a, b))
```

# 4.Convert 2D to 1D

```python
a = np.array([[1, 2], [3, 4], [5, 6]])
b = np.array([[7, 8], [9, 10]])
c = np.vstack((a, b))
c.flatten()
```

# 5.Interchange two axis of an array

```python
d = np.array([[1, 2, 3], [4, 5, 6]])
d.transpose()
```
